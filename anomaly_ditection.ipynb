Question 1 : What is Dimensionality Reduction? Why is it important in machine
learning?


Ans- Dimensionality reduction is the process of reducing the number of features (or variables) in a dataset while keeping the essential information intact. It is important in machine learning because it simplifies models, improves computational efficiency, and helps prevent the "curse of dimensionality" which can degrade performance. By reducing complexity, it leads to faster training times and helps avoid overfitting.

Question 2: Name and briefly describe three common dimensionality reduction
techniques..

Ans- Three common dimensionality reduction techniques are Principal Component Analysis (PCA), Linear Discriminant Analysis (LDA), and Autoencoders.

1. Principal Component Analysis (PCA)
Description: PCA is a linear technique that transforms the data into a new coordinate system where the axes, or "principal components," are ranked by the amount of variance they capture in the data. It identifies the directions of greatest variability and projects the data onto a lower-dimensional subspace defined by the most informative principal components.

2. Linear Discriminant Analysis (LDA)
Description: LDA is a linear technique that is used for classification and dimensionality reduction. It projects data onto a lower-dimensional space in a way that maximizes the distance between different classes while minimizing the variance within each class.

3. Autoencoders
Description: Autoencoders are a type of neural network that are trained to reconstruct their input. They have an "encoder" that compresses the data into a lower-dimensional "bottleneck" layer, and a "decoder" that tries to reconstruct the original input from this compressed representation. The compressed representation in the bottleneck layer is the reduced-dimension version of the data.

Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.


Ans- Clustering in unsupervised learning is a technique for grouping unlabeled data points into clusters based on their similarities or differences. Unlike supervised learning, where the algorithm is trained on labeled data, unsupervised learning algorithms discover inherent patterns and structures within the data without prior knowledge of categories or classes. The goal of clustering is to find natural groupings in the data, where data points within a cluster are more similar to each other than to those in other clusters.

Here are three popular clustering algorithms: k-means clustering.
This algorithm partitions data into a pre-defined number of k clusters. It works by iteratively assigning data points to the closest cluster centroid and then updating the centroid's position to the mean of its assigned points. The process continues until cluster assignments no longer change or a maximum number of iterations is reached. hierarchical clustering.
Hierarchical clustering builds a hierarchy of clusters, either by starting with individual data points and progressively merging them (agglomerative) or by starting with one large cluster and recursively dividing it (divisive). The result is often represented as a dendrogram, which illustrates the merging or splitting process and the relationships between clusters.

Question 4: Explain the concept of anomaly detection and its significance.

Ans-Anomaly detection is the process of identifying rare occurrences, data points, or events that deviate significantly from normal or expected behavior within a dataset

Concept of anomaly detection
Identifying outliers: Anomaly detection systems look for data that doesn't fit the established pattern. These discrepancies are often called "outliers" or "exceptions".
Distinguishing normal from abnormal: A core function is to establish a baseline of "normal" behavior and then identify anything that falls outside that baseline.
Various forms of anomalies: Anomalies can signal a wide range of things, from a system failure to a marketing opportunity. They can result from system failures, human error, malicious activity, or even natural events

Question 5: List and briefly describe three types of anomaly detection techniques.

Ans- Three types of anomaly detection techniques are statistical methods, machine learning models, and proximity-based methods. Statistical methods use predefined rules or thresholds, machine learning models learn patterns from data to identify outliers, and proximity-based methods find anomalies based on their distance or density from other data points.

1. Statistical methods Description: These techniques define "normal" behavior by modeling the statistical properties of the data. Anomalies are identified when data points fall outside of a predefined range, often based on a certain number of standard deviations from the mean.Example: Setting a rule that any transaction over \(\$10,000\) is flagged as a potential anomaly, or using a formula like the Z-score to find points that are unusually far from the average.2. Machine learning models Description: This approach uses algorithms to learn patterns from the data itself. The models can be trained in a supervised (with labeled normal/anomalous data), unsupervised (without labels), or semi-supervised (with a mix of both) manner to determine what is normal and then flag data that deviates.

Question 6: What is time series analysis? Mention two key components of time series data.

Ans-Time series analysis is a statistical technique used to analyze data points recorded over time to identify patterns and make forecasts. Two key components of time series data are trend, which is the long-term upward or downward movement, and seasonality, which is the regular, predictable fluctuation within a year.

Components of time series data
Trend: This is the long-term progression of the data. It shows the general direction, such as an increasing trend in sales over several years or a decreasing trend in a product's market share.
Seasonality: This refers to patterns that repeat over a fixed period, such as a year. Examples include the increase in ice cream sales during summer or higher retail sales during the holiday season.
Cyclical: Similar to seasonality, but on a longer, irregular timescale. This component represents up-and-down movements that are not fixed in duration and can be influenced by business cycles.
Irregular/Random: This component accounts for unpredictable, random variations that cannot be explained by the other components.


Question 7: Describe the difference between seasonality and cyclic behavior in time series.


Ans- The key difference between seasonality and cyclic behavior in time series is that seasonality has a fixed and predictable period tied to the calendar, while cyclic behavior involves fluctuations with a variable, non-fixed period.

Seasonality
Fixed Period: Seasonal patterns repeat over an exact, constant time interval, such as daily, weekly, monthly, or yearly.
Predictable: Because the period is fixed and known, seasonal variations are highly predictable and can be easily modeled and forecasted.
Cause: They are typically driven by calendar-related events, weather, holidays, or social habits (e.g., higher retail sales during the Christmas season, increased ice cream sales in summer).
Duration: The period is usually a year or less (e.g., weekly traffic patterns, monthly sales reports).

Cyclic Behavior
Variable Period: Cyclic patterns exhibit rises and falls that are not of a fixed period or duration.
Less Predictable: The duration and magnitude of cycles can vary, making them much harder to forecast accurately than seasonal patterns.
Cause: They are often influenced by broader, longer-term economic, political, or social factors, such as business cycles (periods of expansion and contraction in an economy) or population dynamics in nature.
Duration: Cycles are typically longer than a year, often lasting several years or even decades (e.g., an economic recession might last 3 years, while the next one could last 5 years).

Question 8: Write Python code to perform K-means clustering on a sample dataset.

Ans-

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs


X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, random_state=42)
kmeans.fit(X)


labels = kmeans.labels_

centroids = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

print("Cluster Labels:\n", labels[:10])
print("\nCluster Centroids:\n", centroids)
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.


Inheritance in Object-Oriented Programming (OOP) is a mechanism where a new class (child or derived class) can inherit attributes and methods from an existing class (parent or base class). This promotes code reusability and establishes a hierarchical relationship between classes, representing an "is-a" relationship (e.g., a "Dog is an Animal").

```python

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def speak(self):
        return f"{self.name} says Woof!"

    def fetch(self):
        return f"{self.name} is fetching the ball."

my_animal = Animal("Generic Animal")
my_dog = Dog("Buddy", "Golden Retriever")

print(my_animal.speak())
print(my_dog.speak())
print(my_dog.fetch())
print(my_dog.name)
```
